**Stock price prediction** is the practice of forecasting future stock values using historical data, statistical models, and machine learning algorithms. This analytical approach helps investors, traders, and financial analysts make informed decisions by identifying patterns, trends, and potential market movements.

### 🔹Key Approaches to Stock Price Prediction:

* **Technical Analysis:** Examines historical price patterns, trading volumes, and chart indicators to predict future movements. Common indicators include Moving Averages, Relative Strength Index (RSI), and Bollinger Bands.

* **Fundamental Analysis:** Evaluates company financials, industry trends, and economic indicators to assess intrinsic stock value. This includes analyzing revenue growth, profit margins, and market position.

* **Machine Learning & AI:** Utilizes advanced algorithms (LSTM, ARIMA, Prophet, Random Forests, etc.) to identify complex patterns in historical data and generate predictions based on multiple variables.

* **Sentiment Analysis:** Analyzes news articles, social media, and market sentiment to gauge investor psychology and potential price impacts.

<br>


### 🔹Why Stock Price Prediction Matters:

* **Investment Strategy:** Helps investors determine optimal entry and exit points for positions, maximizing potential returns.

* **Risk Management:** Enables better assessment of potential losses and supports informed portfolio diversification decisions.

* **Market Understanding:** Provides insights into market dynamics, company performance trends, and sector movements.

* **Financial Planning:** Assists in long-term financial planning, retirement strategies, and wealth management.

<br>

### 🔸Limitations and Considerations:

While stock price prediction models can provide valuable insights, it's important to note that:

- **Market Unpredictability:** Markets are influenced by unforeseen events (geopolitical issues, natural disasters, regulatory changes, etc.)

- **No Guarantees:** Past performance doesn't guarantee future results; historical patterns may not repeat

- **Complementary Tool:** Predictions should be used as one tool among many in investment decision-making, not as sole criteria

- **Accuracy Limits:** No model can predict stock prices with 100% accuracy due to market complexity and randomness

- **External Factors:** Global economic conditions, interest rates, and competitor actions can significantly impact predictions

**In this project**, we will apply various analytical techniques and machine learning models to NVIDIA's historical stock data to identify trends, patterns, and potential future price movements, while acknowledging these inherent limitations.

## **5.2 Daily Returns Over Time**
Daily returns over time in stocks data refers to the **percentage change in the price of a stock from one day to the next**. It's a way to track how much an investment in a particular stock has gained or lost each day.
#### Example:
- Day 1: Stock price is 100
- Day 2: Stock price is 105
- The daily return for Day 2 would be calculated as:
((Day 2 Price - Day 1 Price) / Day 1 Price) * 100 = (($105 - $100) / $100) * 100 = 5%

```python
df['Daily Return'] = df['Close'].pct_change()

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Daily Return'], label='Daily Return', color = 'blue')
plt.title('Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid(False)
plt.show()
```

## **5.3 Volatility**
Volatility in stock data refers to the **degree of variation or fluctuation in a stock's price over a given period**. It essentially measures the magnitude of price swings, regardless of whether they are upward or downward. Higher volatility indicates a greater potential for both large gains and large losses, making the stock riskier.

```python
rolling_windows = {
    'Weekly': 5,
    'Monthly': 21,
    'Yearly': 252
}

# Calculate volatility for each period
for label, window in rolling_windows.items():
    df[f'{label} Volatility'] = df['Daily Return'].rolling(window=window).std()

# Plot each volatility separately
for label in rolling_windows.keys():
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df[f'{label} Volatility'], label=f'{label} Volatility', color='red')
    plt.title(f'{label} Stock Price Volatility')
    plt.xlabel('Date')
    plt.ylabel('Volatility')
    plt.legend()
    plt.grid(False)
    plt.show()
```

```python
# Calculate daily volatility (standard deviation of daily returns)
daily_volatility = np.std(df["Daily Return"])

# Weekly volatility (assuming 5 trading days in a week)
weekly_volatility = daily_volatility * np.sqrt(5)

# Monthly volatility (assuming approximately 21 trading days in a month)
monthly_volatility = daily_volatility * np.sqrt(21)

# Annualized volatility (assuming 252 trading days in a year)
annualized_volatility = daily_volatility * np.sqrt(252)

# Convert to percentages
daily_volatility_percentage = daily_volatility * 100
weekly_volatility_percentage = weekly_volatility * 100
monthly_volatility_percentage = monthly_volatility * 100
annualized_volatility_percentage = annualized_volatility * 100

# Print results in percentage format
print(f"Daily Volatility: {daily_volatility_percentage:.2f}%")
print(f"Weekly Volatility: {weekly_volatility_percentage:.2f}%")
print(f"Monthly Volatility: {monthly_volatility_percentage:.2f}%")
print(f"Annual Volatility: {annualized_volatility_percentage:.2f}%")
```

### **Observation :**
- This stock is risky over any timeframe—weekly, monthly, or yearly. This is because every so often, it makes a huge price jump or drop. Even though the stock is calm on most days, these few dramatic events are so powerful that they define its overall high-risk, high-reward nature.

## **5.4 Moving Average**
Moving average is a statistical method used to analyze data points by creating a series of averages of a fixed size that move over a dataset. This helps to smooth out short-term fluctuations and reveal underlying trends.

In simpler terms, it's like taking a rolling average of a set of numbers. For example, a 50-day moving average would calculate the average price of a stock over the past 50 days, and this average would be updated daily as new data points become available.

This technique is commonly used in financial analysis, particularly for analyzing stock prices. It can help identify trends, support and resistance levels, and generate trading signals.

```python
# Calculate moving average
df['MA_30'] = df['Close'].rolling(window=30).mean()
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()

# Visualize
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA_30'], label='30-day MA', color = 'red')
plt.plot(df['MA_50'], label='50-day MA', color = 'green')
plt.plot(df['MA_200'], label='200-day MA')
plt.title('Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## **6.1 Long Short-Term Memory**

LSTM is a type of recurrent neural network (RNN) architecture designed to model and predict sequences of data. It is particularly effective in capturing long-term dependencies and managing issues related to vanishing gradients that are common in traditional RNNs. LSTMs achieve this through their special gating mechanisms that control the flow of information.


**1. Handling Sequential Data**
Stock prices are inherently sequential data where the value at a given time depends on previous values. LSTMs are designed to handle such sequential dependencies and can learn from patterns over time.

**2. Capturing Long-Term Dependencies**
LSTMs are capable of capturing long-term dependencies in data. In the context of stock prices, this means they can remember information from many time steps ago, which is important for understanding trends and making predictions based on historical data.

**3. Learning Complex Patterns**
LSTMs can model complex patterns in time series data. Stock price movements can be influenced by a variety of factors including past prices, market trends, and external events. LSTMs can capture these complexities better than simpler models due to their advanced memory mechanisms.

**4. Flexibility with Input Features**
LSTMs can be used with various types of input features, not just raw stock prices. This allows analysts to incorporate additional information such as trading volumes, technical indicators, or even external factors like economic indicators or news sentiment.

**5. Improving Forecast Accuracy**
Because LSTMs are designed to learn from sequences, they can often provide more accurate forecasts compared to simpler models that do not account for temporal dependencies. This is crucial for stock price prediction, where accurate forecasting can lead to better trading decisions.

```python
# Function to prepare the data for LSTM
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

# Hyperparameter ranges
look_back_values = [5, 21, 63, 252]
units_values = [50, 100, 150]

best_rmse = float('inf')
best_look_back = None
best_units = None
best_model = None
```

```python
for look_back in look_back_values:
    for units in units_values:
        print(f"\nTraining with look_back={look_back} and units={units}")

        # Create the training and testing datasets
        X_train, Y_train = create_dataset(train_data, look_back)
        X_test, Y_test = create_dataset(test_data, look_back)

        # Reshape the input to be [samples, time steps, features] which is required for LSTM
        X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
        X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

        # Create a single-layer LSTM model
        model_lstm = Sequential()
        model_lstm.add(LSTM(units=units, input_shape=(X_train.shape[1], 1)))
        model_lstm.add(Dense(1))

        # Compile the model
        model_lstm.compile(loss='mean_squared_error', optimizer='adam')

        # Train the model
        model_lstm.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

        # Make predictions
        train_predict = model_lstm.predict(X_train)
        test_predict = model_lstm.predict(X_test)

        # Invert predictions back to original scale
        train_predict = scaler.inverse_transform(train_predict)
        Y_train = scaler.inverse_transform([Y_train])
        test_predict = scaler.inverse_transform(test_predict)
        Y_test = scaler.inverse_transform([Y_test])

        # Evaluate the model
        train_rmse = np.sqrt(mean_squared_error(Y_train[0], train_predict[:,0]))
        test_rmse = np.sqrt(mean_squared_error(Y_test[0], test_predict[:,0]))

        print(f"Train Root Mean Squared Error: {train_rmse}")
        print(f"Test Root Mean Squared Error: {test_rmse}")

        # Check if this is the best model so far
        if test_rmse < best_rmse:
            best_rmse = test_rmse
            best_look_back = look_back
            best_units = units
            best_model = model_lstm
```

```python
print(f"Best look_back: {best_look_back}")
print(f"Best units: {best_units}")
print(f"Best Test RMSE: {best_rmse}")
```

**Holt's Exponential Smoothing**, also known as **Double Exponential Smoothing**, is a time series forecasting technique that extends Simple Exponential Smoothing by adding a trend component. This method is useful for data with a trend but no seasonality. It provides a way to account for both the level and trend in the time series data, which allows for more accurate forecasts when there is a linear trend present.

### **Components of Holt's Exponential Smoothing**
Holt's method involves two main components:

* **Level** (L): The smoothed estimate of the current level of the time series.
* **Trend** (T): The smoothed estimate of the trend in the time series.


### **Key Parameters:**

Holt’s Exponential Smoothing uses two smoothing constants:

**Alpha** (α): The smoothing constant for the level component.

* Range: Between 0 and 1.
* Role: Determines how quickly the model reacts to changes in the level of the time series.

**Beta** (β): The smoothing constant for the trend component.

* Range: Between 0 and 1.
* Role: Determines how quickly the model adapts to changes in the trend of the time series.

## **6.4 Holts-Winter Exponential Smoothing** (Triple Exponential Smoothing)

**Holt-Winters Exponential Smoothing**, also known as **Triple Exponential Smoothing**, is a forecasting technique designed to handle time series data that exhibit trends and seasonality. This method extends Holt’s Exponential Smoothing by incorporating a seasonal component, making it suitable for series with both trend and periodic seasonal fluctuations.

### **Components of Holt-Winters Exponential Smoothing**
The Holt-Winters model includes three components:

**Level** (L): Represents the smoothed estimate of the current level of the time series.

**Trend** (T): Represents the smoothed estimate of the trend in the time series.

**Seasonal Component** (S): Represents the repeating seasonal pattern in the time series.
<br></br>
### **Types of Holt-Winters Models**
**Additive Model**: Used when the seasonal variations are approximately constant throughout the series.

**Multiplicative Model**: Used when the seasonal variations change proportionally with the level of the time series.
<br></br>
### **Key Parameters**
Holt-Winters Exponential Smoothing uses three smoothing constants:

**Alpha** (α): The smoothing constant for the level component.

* Range: Between 0 and 1.
* Role: Determines the weight given to the most recent observation in the level calculation.

**Beta** (β): The smoothing constant for the trend component.

* Range: Between 0 and 1.
* Role: Determines the weight given to the most recent trend estimate.

**Gamma** (γ): The smoothing constant for the seasonal component.

* Range: Between 0 and 1.
* Role: Determines the weight given to the most recent seasonal component.

```python
# Split the data into train and test sets
train, test = train_test_split(df, test_size=0.2, shuffle=False)
print(f"Training Data Size: {len(train)}")
print(f"Testing Data Size: {len(test)}")
```

##       

## **6.4.1 Hyperparameter tuning to obtain the optimal "α", "β" and "γ" values**

```python
# Triple Exponential Smoothing with hyperparameter tuning
best_rmse = float('inf')
best_alpha, best_beta, best_gamma = 0, 0, 0

alpha_values = np.arange(0.1, 1.0, 0.1)
beta_values = np.arange(0.1, 1.0, 0.1)
gamma_values = np.arange(0.1, 1.0, 0.1)

for alpha in alpha_values:
    for beta in beta_values:
        for gamma in gamma_values:
            try:
                model = ExponentialSmoothing(train['Close'], trend='add', seasonal='add', seasonal_periods=5)
                fit = model.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
                predictions = fit.forecast(len(test))

                # Check for NaN values in predictions
                if predictions.isna().any():
                    continue

                # Ensure alignment of forecast and test data
                if len(predictions) != len(test):
                    continue

                rmse = np.sqrt(mean_squared_error(test['Close'], predictions))
                if rmse < best_rmse:
                    best_rmse = rmse
                    best_alpha, best_beta, best_gamma = alpha, beta, gamma
            except Exception as e:
                print(f"Error with alpha={alpha}, beta={beta}, gamma={gamma}: {e}")
```

## **6.6 AutoRegressive Integrated Moving Average** (ARIMA)

**ARIMA** (AutoRegressive Integrated Moving Average) is a widely used statistical model for time series forecasting and analysis. It is designed to handle a variety of time series data, particularly those that exhibit trends and seasonality. The ARIMA model is especially effective for stationary time series data, but it can be adapted to handle non-stationary data through differencing.

### **Components of ARIMA**
ARIMA stands for:

**AutoRegressive** (AR): This component models the relationship between an observation and a number of lagged observations (previous values). It captures the dependency between the current value and its previous values.

**Integrated** (I): This component involves differencing the time series to make it stationary. Stationarity means that the statistical properties of the series (mean, variance) are constant over time.

**Moving Average** (MA): This component models the relationship between an observation and a number of lagged forecast errors. It captures the dependency between the current value and past forecast errors.
<br></br>

### **Key Parameters**
The ARIMA model is defined by three parameters:

* **p**: The number of lag observations included in the model (the order of the AutoRegressive component).
* **d**: The number of times that the raw observations are differenced (the degree of differencing).
* **q**: The size of the moving average window (the order of the Moving Average component).

```python
def test_stationarity(timeseries):

    #Determine rolling statistics
    movingAverage = timeseries.rolling(window=5).mean()
    movingSTD = timeseries.rolling(window=5).std()

    #Plot rolling statistics
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
    std = plt.plot(movingSTD, color = 'black', label = 'Rolling std')

    plt.legend(loc='best')
    plt.title('Rolling Mean')
    plt.show(block=False)

    #Perform Dickey–Fuller test:
    print('Results of Dickey Fuller test:')
    dftest = adfuller(timeseries['Close'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(timeseries):
    # Perform Augmented Dickey-Fuller test
    result = adfuller(timeseries.dropna())
    #print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    #print('Critical Values:')
    #for key, value in result[4].items():
        #print('\t%s: %.3f' % (key, value))

    # Interpret results
    if result[1] <= 0.05:
        print("Reject the null hypothesis (H0)")
        print('Data is stationary.')
    else:
        print("Fail to reject the null hypothesis (H0)")
        print('Data is non-stationary.')
```

```python
# Get the initial value (the last value before the differencing started)
initial_value = df['Close'].iloc[0]

# Convert differenced data back to original scale
def inverse_difference(differenced_data, initial_value):
    return initial_value + np.cumsum(differenced_data)

# Convert train, test, and forecast to original scale
train_original = inverse_difference(train['Close'], initial_value)
test_original = inverse_difference(test['Close'], train_original.iloc[-1])
forecast_original = inverse_difference(forecast, train_original.iloc[-1])

# Create a datetime index for plotting
date_index = pd.date_range(start=df.index[1], periods=len(df)-1, freq='D')

# Create a plot
plt.figure(figsize=(15, 8))

# Plot training data
plt.plot(date_index[:len(train)], train_original, label='Train', color='blue')

# Plot test data
plt.plot(date_index[len(train):], test_original, label='Test', color='green')

# Plot forecast
plt.plot(date_index[len(train):], forecast_original, label='Forecast', color='red', linestyle='-')

plt.title('Stock Price - Original Scale')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(False)

# Rotate and align the tick labels so they look better
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()

# Print RMSE on the original scale
ARIMA_rmse_original = np.sqrt(mean_squared_error(test_original, forecast_original))
print(f"RMSE on Test Set (Original Scale): {ARIMA_rmse_original}")
```
